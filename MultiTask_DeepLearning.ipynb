*Creating the New Target Variable*

We define a new variable (HouseCategory) to classify the houses based on the 'HouseStyle', 'BldgType', 'YearBuilt', and 'Year RemodAdd'
Since the data is not large enough, we do not create a category for each combination of the mentioned variables. Instead, we define classes based on the following rules:

1- If a house has a 'BldgType' of '1Fam' or a 'HouseStyle' of '1Story'/'1.5Fin'/'1.5Unf', it is considered a small house

2- If the 'YearBuilt' for a house is 2000 or newer, the house is considered a new house.

3- If the 'YearRemodAdd' is larger than the 'YearBuilt' by more than 20, the house is considered remodeled.

Finally, the HouseCategory of 0 is assigned to houses that are small and new, 1 is assigned to small, old, remodeled houses, 2 is assigned to small, old, not remodeled houses, 3 is assigned to big and new houses, 4 is assigned to big, old, remodeled houses, and 5 is assigned to big, old, not remodeled houses. It is assumed that houses built after 2000 are not remodeled and are all considered new.

```python
train['small'] = np.where((train['BldgType'] == '1Fam') | (train['HouseStyle'] == '1Story') | (train['HouseStyle'] == '1.5Fin') | (train['HouseStyle'] == '1.5Unf'), 'Yes', 'No')
train['new'] = np.where(train['YearBuilt'] >= 2000, 'Yes', 'No')
train['remodeled'] = np.where(train['YearRemodAdd'] >= train['YearBuilt']+20, 'Yes', 'No')
```

```python
train['HouseCategory'] = 'def'
```

```python
train['HouseCategory'] = np.where((train['small'] == 'Yes') & (train['new'] == 'Yes'), 0, train['HouseCategory'])
train['HouseCategory'] = np.where((train['small'] == 'Yes') & (train['new'] == 'No') & (train['remodeled'] == 'Yes'), 1, train['HouseCategory'])
train['HouseCategory'] = np.where((train['small'] == 'Yes') & (train['new'] == 'No') & (train['remodeled'] == 'No'), 2, train['HouseCategory'])
train['HouseCategory'] = np.where((train['small'] == 'No') & (train['new'] == 'Yes'), 3, train['HouseCategory'])
train['HouseCategory'] = np.where((train['small'] == 'No') & (train['new'] == 'No') & (train['remodeled'] == 'Yes'), 4, train['HouseCategory'])
train['HouseCategory'] = np.where((train['small'] == 'No') & (train['new'] == 'No') & (train['remodeled'] == 'No'), 5, train['HouseCategory'])
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
```

```python
class MultiTaskModel(pl.LightningModule):
    def __init__(self):
        super(MultiTaskModel, self).__init__()
        self.shared_layers = nn.Sequential(
            nn.Linear(73, 36),
            nn.ReLU(),
            nn.Linear(36, 18),
            nn.ReLU(),
        )
        self.price_head = nn.Sequential(
            nn.Linear(18, 10),
            nn.ReLU(),
            nn.Linear(10, 1),
        )
        self.category_head = nn.Sequential(
            nn.Linear(18, 10),
            nn.ReLU(),
            nn.Linear(10, 6),
        )

        nn.init.xavier_normal_(self.shared_layers[0].weight)
        nn.init.xavier_normal_(self.price_head[0].weight)
        nn.init.xavier_normal_(self.category_head[0].weight)

        self.price_loss = nn.MSELoss()
        self.category_loss = nn.CrossEntropyLoss()

    def forward(self, x):
        x = self.shared_layers(x)
        price = self.price_head(x)
        category = self.category_head(x)
        return price, category

    def training_step(self, batch, batch_idx):
        x, price_targets, category_targets = batch
        price_pred, category_pred = self.forward(x)
        price_loss = self.price_loss(price_pred.squeeze(), price_targets)
        category_loss = self.category_loss(category_pred, category_targets)
        loss = price_loss + category_loss
        self.log('train_loss', loss)
        return loss

    def validation_step(self, batch, batch_idx):
        x, price_targets, category_targets = batch
        price_pred, category_pred = self.forward(x)
        price_loss = self.price_loss(price_pred.squeeze(), price_targets)
        category_loss = self.category_loss(category_pred, category_targets)
        loss = price_loss + category_loss
        self.log('val_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=0.001)
        return optimizer

```

```python
class HouseDataset(Dataset):
    def __init__(self, features, price_targets, category_targets):
        self.features = features
        self.price_targets = price_targets
        self.category_targets = category_targets

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        x = torch.tensor(self.features.iloc[idx].values, dtype=torch.float32)
        price_target = torch.tensor(self.price_targets.iloc[idx], dtype=torch.float32)
        category_target = torch.tensor(self.category_targets.iloc[idx], dtype=torch.long)
        return x, price_target, category_target
```

```python
features = X_train
price_targets = y_train['SalePrice']
category_targets = y_train['HouseCategory']
```

```python
dataset = HouseDataset(features, price_targets, category_targets)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
model = MultiTaskModel()
trainer = pl.Trainer(max_epochs=20)
trainer.fit(model, dataloader)
```
